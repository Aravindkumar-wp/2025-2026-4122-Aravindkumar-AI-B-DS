```python
import numpy as np
import math
population=np.array([10,20,30,40,50])
N=len(population)
print("POPULATION SIZE(N):",N)
mu=np.mean(population)
print("POPULATION MEAN(M):",mu)
sigma2=np.var(population,ddof=1)
print("POPULATION VARIANCE(S^2):",sigma2)
sigma=np.sqrt(sigma2)
print("POPULATION STANDARD DEVIATION(S):",sigma)
sample=np.array([20,30,40,50])
n=len(sample)
print("\n SAMPLE SIZE:",n)
x_bar=np.mean(sample)
print("SAMPLE MEAN:",x_bar)
s2=np.var(sample,ddof=1)
print("SAMPLE VARIANCE:",s2)
s=np.sqrt(s2)
print("SAMPLE SD:",s)
SE=sigma/math.sqrt(n)
print("STANDARD ERROR:",SE)
Z=(x_bar-mu)/SE
print("\n Z-SCORE:",Z)
```
