```python
# z_score
import numpy as np
data=[10,20,30,40,50]
mean=np.mean(data)
print("MEAN:",mean)
std_dev=np.std(data,ddof=1)
print("STANDARD DEVIATION:",std_dev)
x=40
z_score=(x-mean)/std_dev
print("Z_SCORE:",z_score)
z_score=[(x-mean)/std_dev for x in data]
print("Z_SCORE:",z_score)
```
