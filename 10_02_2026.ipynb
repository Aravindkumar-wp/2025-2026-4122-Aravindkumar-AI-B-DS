```python
#5Q.
# Linear Search

n = int(input("enter no. of elements:"))
arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

key = int(input("enter element to search:"))

found = False

for i in range(n):
    if arr[i] == key:
        print(f"element {key} found at position {i+1}")
        found = True
        break

if not found:
    print("element not found:")
```

```python
#Binary Search

n = int(input("enter number of elements:"))

arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

arr.sort()
print("\n sorted list :", arr)

key = int(input("enter element to search:"))

low = 0
high = n-1

found = False
while low <= high:
    mid = (low + high) // 2

    if arr[mid] == key:
        print(f"element {key} found at position {mid+1}")
        found = True
        break

    elif key < arr[mid]:
        high = mid - 1
    else:
        low = mid + 1

if not found:
    print(f"element {key} not found in the list")
```

```python
#6Q.
# Bubble sort
n = int(input("enter num of elements:"))

arr = []

for i in range(n):
    arr.append(int(input(f"enter element {i+1}:")))

print("\n Original array:", arr)

# Bubble sort logic
for i in range(n-1):
    for j in range(n-1-i):
        if arr[j] > arr[j+1]:
            # Swap
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("Sorted array:", arr)
```

```python
#selection sort

n=int(input("enter no.of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"enter elements{i+1}")))

print("\n Original array:",arr)
# selection sort logic
for i in range(n-1):

  min_index=i
  for j in range(i+1,n):
    if arr[j]<arr[min_index]:
      min_index=j
  #swap found min element with current element 'i'
  if min_index != i:
    arr[i],arr[min_index]=arr[min_index],arr[i]
print("\n Sorted array:",arr)
```

```python
import pandas as pd
data={
      "ID":[101,102,103,104,105],
      "Name":["Alice","Bob","Charlie","David","Eve"],
      "Age":[22,25,23,24,26],
      "Marks":[85,78,92,88,90]
}
df=pd.DataFrame(data)
print("Original dataframe:")
print(df)
print("\nStudents with marks>85:")
print(df[df["Marks"]>85])
print("\n Students with age<25:")
print(df[df["Age"]<25])
# Students with age > 23 AND marks >= 88
print("\nStudents with age > 23 and marks >= 88:")
print(df[(df["Age"] > 23) & (df["Marks"] >= 88)])

# Students with marks > 90 OR age < 23
print("\nStudents with marks > 90 or age < 23:")
print(df[(df["Marks"] > 90) | (df["Age"] < 23)])
```

```python
# Data Manipulation
import pandas as pd

# create Dataframe
data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 22],
    "Marks": [70, 80, 90]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# Add new column
df["Bonus"] = 5
print("\n After adding Bonus:")
print(df)

# Increase marks
df["Marks"] = df["Marks"] + 2
print("\n After increasing marks")
print(df)

# Change value
df.loc[0, "Age"] = 21
print("\n After changing age of first student:")
print(df)

# Rename column
df.rename(columns={"total": "final"}, inplace=True)
print("\n After renaming total to final:")
print(df)

# Remove column
df.drop("Bonus", axis=1, inplace=True)
print("\n After removing Bonus:")
print(df)
```
